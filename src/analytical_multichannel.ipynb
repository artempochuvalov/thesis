```python
import numpy as np
from math import factorial, pow
from numba import njit, prange
from tqdm.notebook import trange
```

```python
def get_right_denominator(
    n: int,
    N: int,
    alpha: float,
    betta: float
) -> float:
    sum = 0.
    for m in prange(1, N + 1):
        denominator = 1.
        for s in range(1, m + 1):
            denominator *= float(n) + float(s) * betta
        sum += np.longlong(alpha) ** m / denominator

    return ((np.longlong(alpha) ** n) / factorial(n)) * sum

def get_left_denominator(
    n: int,
    alpha: float
) -> float:
    sum = 0.
    for m in range(n + 1):
        sum += (np.longlong(alpha) ** m) / factorial(m)

    return sum

def get_numerator(i: int, n: int, alpha: float, betta: float):
    denominator = 1.
    if i <= n:
        denominator = factorial(i)
    else:
        demominator = factorial(n)
        for s in range(1, i - n + 1):
            demominator *= float(n) + float(s) * betta

    result = np.longlong(alpha) ** i
    result= result / denominator

    print(result)
            
    return result
```

```python
def get_probabilities(
    n: int,
    N: int,
    lambda_val: float,
    mu: float,
    gamma: float
) -> float:
    alpha = lambda_val / mu
    betta = gamma / mu

    probabilities = np.empty(N + n + 1)

    denominator = get_right_denominator(n, N, alpha, betta) + get_left_denominator(n, alpha)
    for i in range(probabilities.size):
        numerator = get_numerator(i, n, alpha, betta)
        probabilities[i] = numerator / denominator

    return probabilities
```
